```python
# Import necessary libraries
import json      # For working with JSON data (e.g., parsing or dumping strings)
import ast       # For safely evaluating Python literals (like converting stringified dicts/lists to Python objects)
from pandas import DataFrame, Series  # Import DataFrame and Series directly from pandas for data manipulation
```

```python
import json
import ast

from pandas import DataFrame, Series

# Update this path to point to the actual file on your system
input_path = "/Users/daniel/Documents/Northwestern/MSDS 420 - Database Systems/Final Project Implementation/Final_Project_Phase_1/Amazon/meta_Appliances.json"
output_path = "/Users/daniel/Documents/Northwestern/MSDS 420 - Database Systems/Final Project Implementation/Final_Project_Phase_1/Amazon/Amazon_Appliances_Metadata.json"

# Open the source metadata file (contains one Python-style dictionary per line)
with open(input_path, "r") as fr, open(output_path, "w") as fw:
    for line in fr:
        # Convert the string to a Python dict and then to valid JSON format
        json_dat = json.dumps(ast.literal_eval(line))
        dict_dat = json.loads(json_dat)

        # Write the JSON object to the output file as one line (JSON Lines format)
        json.dump(dict_dat, fw)
        fw.write("\n")

print("Metadata conversion complete.")
```

```python
import json
import ast

# Set full file paths to avoid FileNotFoundError
input_path = "/Users/daniel/Documents/Northwestern/MSDS 420 - Database Systems/Final Project Implementation/Final_Project_Phase_1/Amazon/reviews_Appliances.json"
output_path = "/Users/daniel/Documents/Northwestern/MSDS 420 - Database Systems/Final Project Implementation/Final_Project_Phase_1/Amazon/Amazon_Appliances_Reviews.json"

# Open the original reviews file and prepare to write the cleaned JSON lines
with open(input_path, "r") as fr, open(output_path, "w") as fw:
    for line in fr:
        # Convert the stringified Python dictionary to a proper JSON string
        json_dat = json.dumps(ast.literal_eval(line))
        
        # Parse it back into a Python dictionary
        dict_dat = json.loads(json_dat)

        # Write the dictionary as a single JSON line
        json.dump(dict_dat, fw)
        fw.write("\n")

print("Reviews conversion complete.")
```

```python
# Load the JSON Lines (one JSON object per line) file into a list of Python dictionaries
# This reads each line, parses it into a dictionary, and stores all records in a list
list_of_appliances_metadata_dict_data = [
    json.loads(line)
    for line in open('/Users/daniel/Documents/Northwestern/MSDS 420 - Database Systems/Final Project Implementation/Final_Project_Phase_1/Amazon/Amazon_Appliances_Metadata.json')
]
```

```python
# Convert the list of parsed metadata dictionaries into a Pandas DataFrame
# This creates a structured tabular view of the metadata file
from pandas import DataFrame

appliances_metadata_df = DataFrame(list_of_appliances_metadata_dict_data)

# Optionally preview the first few rows
appliances_metadata_df.head()
```
